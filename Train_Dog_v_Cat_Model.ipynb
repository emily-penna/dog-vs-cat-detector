# Import Dataset from Roboflow
Dataset can be downloaded [here](https://universe.roboflow.com/emilys-workspace/dog-vs-cat-hxiec/dataset/1) in YOLOv5 format

````python
rf = Roboflow(api_key="---")
project = rf.workspace("emilys-workspace").project("dog-vs-cat-hxiec")
dataset = project.version(1).download("yolov5")
````

Edit data.yaml so that the file paths at the bottom of the file look like this:
```
test: ../test/images
train: Dog-vs-Cat-1/train/images
val: Dog-vs-Cat-1/valid/images
```

# Train Model on Dataset

````python
!yolo task=detect \
mode=train \
model=yolov8s.pt \
data={dataset.location}/data.yaml \
epochs=100 \
imgsz=640
````
